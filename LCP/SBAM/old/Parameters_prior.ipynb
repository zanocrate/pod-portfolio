```r
library("ggplot2")
library("grid")
```

```r
theta<-seq(0,20,0.1)
y<-dgamma(theta,3,2)


gamma.max<-theta[which.max(y)]
#my_grob = grobTree(textGrob("Most probabile value: 1", x=0.12,  y=0.95, hjust=0,
#  gp=gpar(col="firebrick", fontsize=12, fontface="bold")))

ggplot()+
    geom_line(aes(theta,y))+
    geom_vline(xintercept = gamma.max,color="firebrick",linetype="dashed")+
#    annotation_custom(my_grob)+
    labs(y="density",title="Gamma distribution prior for the hyperparameter theta")
   
```
