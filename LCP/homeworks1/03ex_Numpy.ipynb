7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
def prime_numbers_slow(N):
    nums = np.arange(0,N)
    mask = np.ones_like(nums,dtype=bool) # all true: every number is initialized as prime
    for num in nums[2:]:
        #nums[num+1:] % num != 0 # this array is false for every multiple of num: these numbers are not prime
        mask[num+1:] = mask[num+1:] & (nums[num+1:] % num != 0) # this logical and operator filters the mask, setting to false the numbers that are not prime found in this iteration
    
    return nums[mask]


def prime_numbers_eratosthenes(N):
    nums = np.arange(0,N)
    mask = np.ones_like(nums, dtype=bool)
    for num in range(2,int(((N-1)**0.5 + 1))): #stopping at square root of N, since every number after has either been already checked or would be used to check numbers larger than N
        if mask[num]: #is this number a prime? if so, use it to mark out other numbers...
            i = 0
            while num*(num+i) <= (N-1): # N-1 because arange generates number up to N excluded
                mask[num*(num+i)]=False
                i+=1
    return nums[mask]

print("Testing both algorithms for N = 100")
print(prime_numbers_slow(100))
print(prime_numbers_eratosthenes(100))

import time
Ns = [100,250,500,750,1000,1500,2000]
T_slow = []
T_erat = []
for N in Ns:
    print("Slow (first) and Erathostenes (second) algorithm CPU time for N={}:".format(N))
    tic = time.time()
    prime_numbers_slow(N)
    toc = time.time()
    T_slow.append(toc-tic) #seconds
    print(T_slow[-1])

    tic = time.time()
    prime_numbers_eratosthenes(N)
    toc = time.time()
    T_erat.append(toc-tic)
    print(T_erat[-1])

fig,ax = plt.subplots()
ax.plot(Ns,T_slow,color='darkgreen',label='No Eratosthenes')
ax.plot(Ns,T_erat,color='darkmagenta', label='Eratosthenese')
ax.set_ylabel('Time (s)')
ax.set_xlabel('N')
ax.legend()
plt.show()


````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
import numpy.random as nr 
nr.seed(1234)
N_walkers,N_steps = 1000, 200
walkers = np.zeros(N_walkers).reshape((-1,1))
for step in range(N_steps):
    steps=nr.randint(2,size=N_walkers)*2 - 1
    walkers=np.hstack((walkers,(walkers[:,-1]+steps).reshape((-1,1))))

# plotting walkers for fun
import matplotlib.pyplot as plt
%matplotlib inline
fig,ax=plt.subplots()
for walker in walkers:
    ax.plot(walker,alpha=0.2)
ax.set_ylabel('Units of distance')
ax.set_xlabel('Timestep')
plt.show()
# plotting average distance walked
distance_sqrd=np.mean(walkers**2,axis=0)
fig,ax=plt.subplots()
ax.plot(distance_sqrd**0.5,label='Average distance walked')
ax.plot(np.sqrt(np.arange(0,200)),label='Sqrt(t)',linestyle='--',alpha=0.5)
ax.set_ylabel('Units of distance')
ax.set_xlabel('Timestep')
ax.legend()
plt.show()
````
